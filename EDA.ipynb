```python
import pandas as pd 
import matplotlib.pyplot as plt
df= pd.read_csv('Cleaned_Dataset.csv')
df

# Basic statitics
df.describe()

# identifying the monthly trends 
df["Date"].value_counts()

df["PaymentMethod"].value_counts()

# Order status distribution
df["OrderStatus"].value_counts()

# Referral source
df["ReferralSource"].value_counts()

# Average total price by payment method
df.groupby("PaymentMethod")["TotalPrice"].mean()


#  identify outliners 
Q1 = df["TotalPrice"].quantile(0.25)
Q3 = df["TotalPrice"].quantile(0.75)

IQR = Q3 - Q1

outliers = df[
    (df["TotalPrice"] < Q1 - 1.5 * IQR) |
    (df["TotalPrice"] > Q3 + 1.5 * IQR)
]

print(outliers)


# Distribution of TotalPrice
plt.hist(df["TotalPrice"], bins=20)
plt.title("Distribution of Total Price")
plt.xlabel("Total Price")
plt.ylabel("Frequency")
plt.show()



```
